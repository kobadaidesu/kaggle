```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('dark')
sns.set_palette('Set2')
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, cross_validate, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import make_scorer, precision_score, recall_score
```

```python
train_data = pd.read_csv('../input/train.csv')
test_data = pd.read_csv('../input/test.csv')
data = pd.concat([train_data, test_data]).reset_index().drop(['index'], axis=1)

str_cols = data.select_dtypes(include=["object"]).columns
data[str_cols] = (data[str_cols]
                .apply(lambda s: s.str.strip())
                .replace(r"^\s*$", np.nan, regex=True))

# Kaggle互換の Survived が欲しい場合（Perished があるときだけ）
if "Perished" in data and "Survived" not in data:
    data["Survived"] = 1 - data["Perished"]
```

```python
data['Surname'] = data['Name'].apply(lambda x: x.split(',')[0])
```

```python
# New Ticket_id column
data['Ticket_id'] = 'new_col'
# Initialize Ticket_id = Pclass + Ticket + Fare + Embarked
def ticket_id(row):
    row['Ticket_id'] = str(row.Pclass) + '-' + str(row.Ticket)[:-1] + '-' + str(row.Fare) + '-' + str(row.Embarked)
    return row

data = data.apply(ticket_id, axis='columns')
# New Group_id column
data['Group_id'] = 'new_col2'
# Initialize Group_id = Surname + Ticket_id
def group_id(row):
    row['Group_id'] = str(row.Surname) + '-' + str(row.Ticket_id)
    return row

data = data.apply(group_id, axis='columns')
```

# Male Model Training

```python
def train_male_knn_model(train_male, features, n_splits=15, random_state=42):
    """
    男性モデルの訓練と予測を行う
    
    Parameters:
    -----------
    train_male : DataFrame
        訓練データ（男性のみ）
    features : list
        使用する特徴量のリスト
    n_splits : int
        クロスバリデーションの分割数
    random_state : int
        乱数シード
    
    Returns:
    --------
    models : list
        訓練済みKNNモデルのリスト [(k, model), ...]
    cv_scores : dict
        クロスバリデーションスコア {k: {'precision': float, 'recall': float}}
    """
    y_m = train_male['Survived']
    X_m = train_male[features]
    
    # StratifiedKFold with random_state for reproducibility
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)
    
    cv_scores = {}
    models = []
    
    for k in range(1, 18):
        # モデル作成
        model = KNeighborsClassifier(n_neighbors=k)
        
        # データの前処理
        X_processed = X_m.copy()
        
        # Pfare の欠損値補完と標準化
        imputer = SimpleImputer()
        X_processed['Pfare'] = imputer.fit_transform(X_processed[['Pfare']])
        scaler = StandardScaler()
        X_processed['Pfare'] = scaler.fit_transform(X_processed[['Pfare']])
        
        # Embarked の欠損値補完（最頻値）
        X_processed['Embarked'] = X_processed['Embarked'].fillna(X_processed['Embarked'].mode()[0])
        
        # カテゴリカル変数をダミー変数に変換
        X_processed = pd.get_dummies(X_processed, columns=['Pclass', 'Embarked'], drop_first=False)
        
        # CVスコア計算
        precision = cross_val_score(model, X_processed, y_m, cv=skf, n_jobs=-1, scoring='precision').mean()
        recall = cross_val_score(model, X_processed, y_m, cv=skf, n_jobs=-1, scoring='recall').mean()
        
        cv_scores[k] = {'precision': precision, 'recall': recall}
        models.append((k, model))
    
    return models, cv_scores

def create_male_ensemble(train_male, test_male, features, k_list=[1, 3, 7], n_splits=15, random_state=42):
    """
    男性用のアンサンブルモデルを作成し、予測を行う
    
    Parameters:
    -----------
    train_male : DataFrame
        訓練データ（男性のみ）
    test_male : DataFrame
        テストデータ（男性のみ）
    features : list
        使用する特徴量のリスト
    k_list : list
        アンサンブルに使用するkの値のリスト
    n_splits : int
        クロスバリデーションの分割数
    random_state : int
        乱数シード
    
    Returns:
    --------
    predictions : array
        テストデータに対する予測結果
    cv_results : dict
        クロスバリデーション結果
    """
    y_m = train_male['Survived']
    X_m = train_male[features].copy()
    X_test_m = test_male[features].copy()
    
    # データの前処理（訓練データ）
    imputer = SimpleImputer()
    X_m['Pfare'] = imputer.fit_transform(X_m[['Pfare']])
    scaler = StandardScaler()
    X_m['Pfare'] = scaler.fit_transform(X_m[['Pfare']])
    X_m['Embarked'] = X_m['Embarked'].fillna(X_m['Embarked'].mode()[0])
    X_m = pd.get_dummies(X_m, columns=['Pclass', 'Embarked'], drop_first=False)
    
    # データの前処理（テストデータ）
    X_test_m['Pfare'] = imputer.transform(X_test_m[['Pfare']])
    X_test_m['Pfare'] = scaler.transform(X_test_m[['Pfare']])
    X_test_m['Embarked'] = X_test_m['Embarked'].fillna(X_test_m['Embarked'].mode()[0])
    X_test_m = pd.get_dummies(X_test_m, columns=['Pclass', 'Embarked'], drop_first=False)
    
    # テストデータに訓練データと同じカラムを持たせる
    for col in X_m.columns:
        if col not in X_test_m.columns:
            X_test_m[col] = 0
    X_test_m = X_test_m[X_m.columns]
    
    # アンサンブルモデル作成
    estimators = [(f'm{k}', KNeighborsClassifier(n_neighbors=k)) for k in k_list]
    ensemble = VotingClassifier(estimators)
    
    # StratifiedKFold with random_state for reproducibility
    skf_male = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)
    
    # クロスバリデーション
    precision = cross_val_score(ensemble, X_m, y_m, cv=skf_male, n_jobs=-1, scoring='precision').mean()
    recall = cross_val_score(ensemble, X_m, y_m, cv=skf_male, n_jobs=-1, scoring='recall').mean()
    accuracy = cross_val_score(ensemble, X_m, y_m, cv=skf_male, n_jobs=-1).mean()
    
    cv_results = {
        'precision': precision,
        'recall': recall,
        'accuracy': accuracy
    }
    
    # モデル訓練と予測
    ensemble.fit(X_m, y_m)
    predictions = ensemble.predict(X_test_m)
    
    return predictions, cv_results
```

```python
cols = ['PassengerId', 'Name', 'Pfare', 'Pclass', 'Embarked']
features = ['Pfare', 'Pclass', 'Embarked']

# ハイパーパラメータチューニング
models, cv_scores = train_male_knn_model(train_male, features, n_splits=15, random_state=42)

precision_m = [cv_scores[k]['precision'] for k in range(1, 18)]
recall_m = [cv_scores[k]['recall'] for k in range(1, 18)]
    
k_range = range(1,18)
plt.figure(figsize=(7,5))
plt.plot(k_range, precision_m, label='15-fold precision')
plt.plot(k_range, recall_m, label='15-fold recall')
plt.axhline(y=0.5, color='r')
plt.xlabel('Value of k for KNN')
plt.title('Precision and recall by number of neighbors', fontsize=14)
plt.legend()
plt.show()
```

```python
# アンサンブルモデルで予測
predictions_m, cv_results_m = create_male_ensemble(
    train_male, test_male, features, 
    k_list=[1, 3, 7], 
    n_splits=15, 
    random_state=42
)

print('15-fold precision of the ensemble: {:.3f}'.format(cv_results_m['precision']))
print('15-fold recall of the ensemble: {:.3f}'.format(cv_results_m['recall']))
print('15-fold accuracy of the ensemble: {:.3f}'.format(cv_results_m['accuracy']))

print('\nThe following 9 adult males are predicted to live:')
test_male.loc[(predictions_m==1), cols]
```

```python
def train_female_knn_model(train_female, features, n_splits=9, random_state=42):
    """
    女性モデルの訓練と予測を行う
    
    Parameters:
    -----------
    train_female : DataFrame
        訓練データ（女性のみ）
    features : list
        使用する特徴量のリスト
    n_splits : int
        クロスバリデーションの分割数
    random_state : int
        乱数シード
    
    Returns:
    --------
    models : list
        訓練済みKNNモデルのリスト [(k, model), ...]
    cv_scores : dict
        クロスバリデーションスコア {k: {'precision': float, 'recall': float}}
    """
    # We set zero_division=0 to avoid raising errors
    custom_precision = make_scorer(precision_score, pos_label=0, zero_division=0)
    custom_recall = make_scorer(recall_score, pos_label=0)
    
    y_f = train_female['Survived']
    X_f = train_female[features]
    
    # StratifiedKFold with random_state for reproducibility
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)
    
    cv_scores = {}
    models = []
    
    for k in range(1, 18):
        # モデル作成
        model = KNeighborsClassifier(n_neighbors=k)
        
        # データの前処理
        X_processed = X_f.copy()
        
        # Pfare の欠損値補完と標準化
        imputer = SimpleImputer()
        X_processed['Pfare'] = imputer.fit_transform(X_processed[['Pfare']])
        scaler = StandardScaler()
        X_processed['Pfare'] = scaler.fit_transform(X_processed[['Pfare']])
        
        # Embarked の欠損値補完（最頻値）
        X_processed['Embarked'] = X_processed['Embarked'].fillna(X_processed['Embarked'].mode()[0])
        
        # カテゴリカル変数をダミー変数に変換
        X_processed = pd.get_dummies(X_processed, columns=['Pclass', 'Embarked'], drop_first=False)
        
        # CVスコア計算
        precision = cross_val_score(model, X_processed, y_f, cv=skf, n_jobs=-1, scoring=custom_precision).mean()
        recall = cross_val_score(model, X_processed, y_f, cv=skf, n_jobs=-1, scoring=custom_recall).mean()
        
        cv_scores[k] = {'precision': precision, 'recall': recall}
        models.append((k, model))
    
    return models, cv_scores

def create_female_ensemble(train_female, test_female, features, k_list=[4, 9, 11], n_splits=9, random_state=42):
    """
    女性用のアンサンブルモデルを作成し、予測を行う
    
    Parameters:
    -----------
    train_female : DataFrame
        訓練データ（女性のみ）
    test_female : DataFrame
        テストデータ（女性のみ）
    features : list
        使用する特徴量のリスト
    k_list : list
        アンサンブルに使用するkの値のリスト
    n_splits : int
        クロスバリデーションの分割数
    random_state : int
        乱数シード
    
    Returns:
    --------
    predictions : array
        テストデータに対する予測結果
    cv_results : dict
        クロスバリデーション結果
    """
    # We set zero_division=0 to avoid raising errors
    custom_precision = make_scorer(precision_score, pos_label=0, zero_division=0)
    custom_recall = make_scorer(recall_score, pos_label=0)
    
    y_f = train_female['Survived']
    X_f = train_female[features].copy()
    X_test_f = test_female[features].copy()
    
    # データの前処理（訓練データ）
    imputer = SimpleImputer()
    X_f['Pfare'] = imputer.fit_transform(X_f[['Pfare']])
    scaler = StandardScaler()
    X_f['Pfare'] = scaler.fit_transform(X_f[['Pfare']])
    X_f['Embarked'] = X_f['Embarked'].fillna(X_f['Embarked'].mode()[0])
    X_f = pd.get_dummies(X_f, columns=['Pclass', 'Embarked'], drop_first=False)
    
    # データの前処理（テストデータ）
    X_test_f['Pfare'] = imputer.transform(X_test_f[['Pfare']])
    X_test_f['Pfare'] = scaler.transform(X_test_f[['Pfare']])
    X_test_f['Embarked'] = X_test_f['Embarked'].fillna(X_test_f['Embarked'].mode()[0])
    X_test_f = pd.get_dummies(X_test_f, columns=['Pclass', 'Embarked'], drop_first=False)
    
    # テストデータに訓練データと同じカラムを持たせる
    for col in X_f.columns:
        if col not in X_test_f.columns:
            X_test_f[col] = 0
    X_test_f = X_test_f[X_f.columns]
    
    # アンサンブルモデル作成
    estimators = [(f'f{k}', KNeighborsClassifier(n_neighbors=k)) for k in k_list]
    ensemble = VotingClassifier(estimators)
    
    # StratifiedKFold with random_state for reproducibility
    skf_female = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)
    
    # クロスバリデーション
    cv_results_detailed = cross_validate(
        ensemble, X_f, y_f, cv=skf_female,
        scoring={
            'precision': custom_precision,
            'recall': custom_recall,
            'accuracy': 'accuracy'
        },
        return_train_score=True
    )
    
    cv_results = {
        'test_precision': cv_results_detailed['test_precision'].mean(),
        'test_precision_std': cv_results_detailed['test_precision'].std(),
        'test_recall': cv_results_detailed['test_recall'].mean(),
        'test_recall_std': cv_results_detailed['test_recall'].std(),
        'test_accuracy': cv_results_detailed['test_accuracy'].mean(),
        'test_accuracy_std': cv_results_detailed['test_accuracy'].std(),
        'train_precision': cv_results_detailed['train_precision'].mean(),
        'train_recall': cv_results_detailed['train_recall'].mean(),
        'train_accuracy': cv_results_detailed['train_accuracy'].mean()
    }
    
    # モデル訓練と予測
    ensemble.fit(X_f, y_f)
    predictions = ensemble.predict(X_test_f)
    
    return predictions, cv_results
```
